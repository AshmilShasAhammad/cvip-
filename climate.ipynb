## Visualizing the data

```python
df['DATE']=pd.to_datetime(df['DATE'])
df['Date'] = df['DATE'].dt.date
df['Time'] = df['DATE'].dt.time
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])


df['Months'] = df['Date'].dt.month_name()
df
```

```python
df['HourlyWetBulbTemperature'].unique
```

## Hourly Wet Bulb Temprature  Daily Analysis

```python
plt.figure(figsize=(10, 6))
plt.plot(df['HourlyWetBulbTemperature'], label='Hourly Wet Bulb Temperature (Celsius)')
plt.xlabel('Years')
plt.ylabel('Temperature (Celsius)')
plt.title('Wet Bulb Temperature Trends Over Time')
plt.legend()
plt.show()
```

## Hourly Wind Speed

```python
plt.figure(figsize=(10, 6))
plt.plot(df['HourlyWindSpeed'], label='Hourly Windspeed')
plt.xlabel('Year')
plt.ylabel('Speed')
plt.title('Hourly Wind Speed')
plt.legend()
plt.show()
```

## Relative Humidity vs Visibility

```python
sns.scatterplot(x='HourlyRelativeHumidity', y='HourlyVisibility', data=df)
plt.title('Relative Humidity vs Visibility')
plt.xlabel('HourlyRelativeHumidity')
plt.ylabel('HourlyVisibility')
plt.show()
```

## Hourly Relative Humidity

```python
plt.figure(figsize=(10, 6))
plt.plot(df['HourlyRelativeHumidity'], linestyle='-', color='purple',label='Hourly Relative Humidity')
plt.xlabel('Year')
plt.ylabel('Humidity')
plt.title('Hourly Relative Humidity Over Time')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Monthly Analysis

Windspeed Trends over Time

```python
plt.figure(figsize=(16, 8))
sns.lineplot(x='Months', y='HourlyWindSpeed', data=df, marker='o', color='g', linestyle='--', markersize=10)

# Add grid and customize labels
plt.grid(True, linestyle='--', alpha=0.9)
plt.xlabel('Months')
plt.ylabel('HourlyWindSpeed')
plt.title('Wind Speed Trends Over Months')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Months', y='HourlyWetBulbTemperature', data=df, marker='o', color='purple', linestyle='-', markersize=10, ci='sd')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Months')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Trends Over Months')
sns.set(style="whitegrid")
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.lineplot(x='Months', y='HourlyRelativeHumidity', data=df, marker='o', color='k', linestyle='--', markersize=10, ci='sd')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Months')
plt.ylabel('Hourly Relative Humidity')
plt.title('Hourly Relative Humidity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
